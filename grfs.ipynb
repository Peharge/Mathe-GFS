```python
import matplotlib.pyplot as plt
import networkx as nx

def plot_neural_network():
    # Erstelle ein neues Diagramm
    G = nx.DiGraph()

    # Definiere Schichten des Netzwerks
    input_layer = ['x1', 'x2', 'x3']  # Eingabeschicht (3 Neuronen)
    hidden_layer = ['h1', 'h2', 'h3', 'h4']  # Verborgene Schicht (4 Neuronen)
    output_layer = ['y']               # Ausgabeschicht (1 Neuron)

    # Füge Knoten (Neuronen) zum Diagramm hinzu
    for node in input_layer + hidden_layer + output_layer:
        G.add_node(node)

    # Verbinde die Neuronen mit den Schichten (Eingabe -> Verborgene Schicht -> Ausgabe)
    for input_node in input_layer:
        for hidden_node in hidden_layer:
            G.add_edge(input_node, hidden_node)

    for hidden_node in hidden_layer:
        G.add_edge(hidden_node, output_layer[0])

    # Positioniere die Knoten in einem Layout
    pos = {
        'x1': (0, 3), 'x2': (0, 2), 'x3': (0, 1),  # Eingabeschicht (3 Neuronen)
        'h1': (1, 3), 'h2': (1, 2.5), 'h3': (1, 2), 'h4': (1, 1.5),  # Verborgene Schicht (4 Neuronen)
        'y': (2, 2)  # Ausgabeschicht (1 Neuron)
    }

    # Zeichne das Netzwerk
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold', edge_color='gray')

    # Zeige das Netzwerk
    plt.title("Visualisierung eines neuronalen Netzwerks")
    plt.show()

# Neuronales Netzwerk visualisieren
plot_neural_network()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrow, Circle

# Funktion, um die Perceptron-Grafik mit Komponenten zu erstellen
def plot_perceptron():
    fig, ax = plt.subplots(figsize=(12, 6))

    # Positionen der Elemente
    inputs = [(0.2, y) for y in [0.8, 0.6, 0.4]]
    sum_node = (0.6, 0.6)
    activation = (0.8, 0.6)
    output = (1.0, 0.6)

    # Knoten zeichnen
    for x, y in inputs:
        circle = Circle((x, y), 0.03, color='lightblue', ec='black', zorder=2)
        ax.add_patch(circle)
        ax.text(x, y, 'x', fontsize=12, ha='center', va='center', zorder=3)

    # Summationsknoten
    sum_circle = Circle(sum_node, 0.04, color='orange', ec='black', zorder=2)
    ax.add_patch(sum_circle)
    ax.text(*sum_node, '$\Sigma$', fontsize=12, ha='center', va='center', zorder=3)

    # Aktivierungsfunktion
    act_circle = Circle(activation, 0.04, color='yellow', ec='black', zorder=2)
    ax.add_patch(act_circle)
    ax.text(*activation, '$\sigma$', fontsize=12, ha='center', va='center', zorder=3)

    # Outputknoten
    out_circle = Circle(output, 0.03, color='purple', ec='black', zorder=2)
    ax.add_patch(out_circle)
    ax.text(*output, '$\hat{y}$', fontsize=12, ha='center', va='center', zorder=3)

    # Verbindungen zeichnen
    for x, y in inputs:
        ax.arrow(x + 0.03, y, sum_node[0] - x - 0.06, sum_node[1] - y,
                 head_width=0.02, head_length=0.02, fc='black', ec='black', length_includes_head=True)

    ax.arrow(sum_node[0] + 0.04, sum_node[1], activation[0] - sum_node[0] - 0.08, 0,
             head_width=0.02, head_length=0.02, fc='black', ec='black', length_includes_head=True)

    ax.arrow(activation[0] + 0.04, activation[1], output[0] - activation[0] - 0.08, 0,
             head_width=0.02, head_length=0.02, fc='black', ec='black', length_includes_head=True)

    # Beschriftungen hinzufügen
    ax.text(0.2, 0.9, 'Eingabe', fontsize=12, ha='center')
    ax.text(0.6, 0.7, 'Summe', fontsize=12, ha='center')
    ax.text(0.8, 0.7, 'Aktivierungsfunktion', fontsize=12, ha='center')
    ax.text(1.0, 0.7, 'Ausgabe', fontsize=12, ha='center')

    # Achsen ausblenden
    ax.axis('off')
    ax.set_xlim(0, 1.2)
    ax.set_ylim(0.2, 1.0)

    plt.show()

# Plot ausführen
plot_perceptron()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a proper 3D visualization of a Tensor
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Tensor dimensions
depth = 3  # Third dimension (number of "layers")
rows = 4   # First dimension (rows)
cols = 4   # Second dimension (columns)

# Example tensor data (values for each element)
tensor_data = np.array([
    [[1, 4, 6, 10], [4, 41, 2, 51], [6, 2, 7, 9], [10, 51, 6, 15]],
    [[2, 5, 7, 12], [5, 42, 3, 52], [7, 3, 8, 10], [12, 52, 7, 16]],
    [[3, 6, 8, 14], [6, 43, 4, 53], [8, 4, 9, 11], [14, 53, 8, 17]]
])

# Spacing between elements in the 3D grid
spacing = 1.5

# Plot each element in the tensor
for d in range(depth):  # Depth (z-axis)
    for r in range(rows):  # Rows (y-axis)
        for c in range(cols):  # Columns (x-axis)
            x = c * spacing
            y = r * spacing
            z = d * spacing

            # Plot the tensor value as text
            ax.text(x, y, z, str(tensor_data[d, r, c]), color='black', ha='center', va='center', fontsize=10)

            # Add a small cube to represent the tensor element
            ax.bar3d(x - 0.2, y - 0.2, z - 0.2, 0.4, 0.4, 0.4, color='skyblue', edgecolor='black', alpha=0.7)

# Set the axis labels to indicate tensor dimensions
ax.set_xlabel('Dimension #2 (Columns)', labelpad=10)
ax.set_ylabel('Dimension #1 (Rows)', labelpad=10)
ax.set_zlabel('Dimension #3 (Depth)', labelpad=10)

# Set the limits for a clean view
ax.set_xlim(-1, cols * spacing)
ax.set_ylim(-1, rows * spacing)
ax.set_zlim(-1, depth * spacing)

# Remove grid and ticks for a cleaner appearance
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Add a title to clarify the structure
ax.set_title("3D Tensor Visualization with Elements", fontsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_tensor_3d(n, m, p):
    """Erstellt ein 3D-Bild zur Visualisierung eines Tensors der Größe n x m x p.

    Args:
        n (int): Anzahl der Zeilen.
        m (int): Anzahl der Spalten.
        p (int): Anzahl der Tiefenebenen.
    """
    # Daten für den Tensor erzeugen
    tensor = np.random.randint(0, 10, (n, m, p))

    # Abbildung erstellen
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Koordinaten für die Visualisierung vorbereiten
    x, y, z = np.meshgrid(
        np.arange(m),
        np.arange(n),
        np.arange(p)
    )

    # Werte aus dem Tensor und ihre Positionen holen
    for i in range(n):
        for j in range(m):
            for k in range(p):
                # Plot eines Würfels für jeden Wert im Tensor
                ax.bar3d(j, i, k, 0.8, 0.8, 0.8, color=plt.cm.viridis(tensor[i, j, k] / 10))
                ax.text(j, i, k, str(tensor[i, j, k]), color="black", ha='center', va='center')

    # Achsentitel setzen
    ax.set_xlabel('Spalten (m)')
    ax.set_ylabel('Zeilen (n)')
    ax.set_zlabel('Ebenen (p)')

    # Achsenbegrenzungen
    ax.set_xlim(0, m)
    ax.set_ylim(0, n)
    ax.set_zlim(0, p)

    plt.title(f"Tensor (rank 3)", fontsize=40, pad=20)
    plt.tight_layout()
    plt.show()

# Beispielausführung
plot_tensor_3d(4, 5, 3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_tensor_2d(n, m):
    """Erstellt ein 2D-Bild zur Visualisierung eines Tensors der Größe n x m.

    Args:
        n (int): Anzahl der Zeilen.
        m (int): Anzahl der Spalten.
    """
    # Daten für den Tensor erzeugen
    tensor = np.random.randint(0, 10, (n, m))

    # Abbildung erstellen
    fig, ax = plt.subplots(figsize=(10, 8))

    # Abstände zwischen den Blöcken einstellen
    spacing = 1.2  # Abstand zwischen den Blöcken

    # Werte aus dem Tensor und ihre Positionen holen
    for i in range(n):
        for j in range(m):
            # Plot eines Rechtecks für jeden Wert im Tensor
            ax.add_patch(plt.Rectangle((j * spacing, i * spacing), 1, 1,
                                       color=plt.cm.viridis(tensor[i, j] / 10)))
            ax.text(j * spacing + 0.5, i * spacing + 0.5, str(tensor[i, j]),
                    color="black", ha='center', va='center')

    # Achsentitel setzen
    ax.set_xlabel('Spalten (m)')
    ax.set_ylabel('Zeilen (n)')

    # Achsenbegrenzungen
    ax.set_xlim(0, m * spacing)
    ax.set_ylim(0, n * spacing)

    plt.title(f"Matrix (rank 2)", fontsize=50, pad=20)
    plt.tight_layout()
    plt.show()

# Beispielausführung
plot_tensor_2d(3, 5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vector_as_visual(n, m):
    """Erstellt eine Visualisierung eines Vektors (der als Tensor betrachtet wird).

    Args:
        n (int): Anzahl der Zeilen.
        m (int): Anzahl der Spalten.
    """
    # Daten für den Vektor erzeugen
    vector = np.random.randint(0, 10, n * m)

    # Abbildung erstellen
    fig, ax = plt.subplots(figsize=(10, 2))

    # Einen Vektor als Balkendiagramm darstellen
    ax.bar(np.arange(len(vector)), vector, color=plt.cm.viridis(vector / 10))

    # Achsentitel setzen
    ax.set_xlabel('Index')
    ax.set_ylabel('Wert')

    # Titel setzen
    ax.set_title(f"Vektor-Darstellung", fontsize=15)

    plt.tight_layout()
    plt.show()

# Beispielausführung
plot_vector_as_visual(3, 5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vector(n):
    """Erstellt ein 2D-Bild zur Visualisierung eines Vektors der Größe n, wobei jeder Vektorwert durch ein Quadrat dargestellt wird.

    Args:
        n (int): Anzahl der Elemente im Vektor.
    """
    # Erzeuge einen zufälligen Vektor
    vector = np.random.randint(0, 10, n)

    # Abbildung erstellen
    fig, ax = plt.subplots(figsize=(3, n * 2))  # Die Höhe wird durch die Anzahl der Vektorelemente bestimmt

    # Abstände zwischen den Blöcken einstellen
    spacing = 1.2  # Abstand zwischen den Blöcken

    # Plot eines Rechtecks für jedes Element im Vektor
    for i in range(n):
        ax.add_patch(plt.Rectangle((0, i * spacing), 1, 1, color=plt.cm.viridis(vector[i] / 10)))  # 1x1 Quadrat
        ax.text(0.5, i * spacing + 0.5, str(vector[i]), color="black", ha='center', va='center')

    # Achsentitel setzen
    ax.set_xlabel('Vektor-Index')
    ax.set_ylabel('Wert')

    # Achsenbegrenzungen
    ax.set_xlim(0, 1.5)
    ax.set_ylim(0, n * spacing)

    plt.title(f"Vektor (rank 1)", fontsize=50, pad=20)
    plt.tight_layout()
    plt.show()

# Beispielausführung
plot_vector(7)  # Ändere n für unterschiedliche Vektorlängen
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vector(n):
    """Erstellt ein 2D-Bild zur Visualisierung eines Vektors der Größe n, wobei jeder Vektorwert durch ein Quadrat dargestellt wird.

    Args:
        n (int): Anzahl der Elemente im Vektor.
    """
    # Erzeuge einen zufälligen Vektor
    vector = np.random.randint(0, 10, n)

    # Abbildung erstellen (größer als zuvor)
    fig, ax = plt.subplots(figsize=(5, n * 3))  # Vergrößerte Höhe für größere Darstellung

    # Abstände zwischen den Blöcken einstellen
    spacing = 1.2  # Abstand zwischen den Blöcken

    # Plot eines Rechtecks für jedes Element im Vektor
    for i in range(n):
        ax.add_patch(plt.Rectangle((0, i * spacing), 1, 1, color=plt.cm.viridis(vector[i] / 20)))  # 1x1 Quadrat
        ax.text(0.5, i * spacing + 0.5, str(vector[i]), color="black", ha='center', va='center')

    # Achsentitel setzen
    ax.set_xlabel('Vektor-Index')
    ax.set_ylabel('Wert')

    # Achsenbegrenzungen
    ax.set_xlim(0, 1.5)
    ax.set_ylim(0, n * spacing)

    # Setze das gleiche Verhältnis für beide Achsen (1:1 Verhältnis)
    ax.set_aspect('equal', adjustable='box')

    plt.title(f"Skalar (rank 0)", fontsize=50, pad=20)
    plt.tight_layout()
    plt.show()

# Beispielausführung
plot_vector(1)  # Ändere n für unterschiedliche Vektorlängen
```
